```python
from bs4 import BeautifulSoup 
import requests

import pandas as pd
from pandas import Series, DataFrame
```

```python
url = 'http://www.ucop.edu/operating-budget/budgets-and-reports/legislative-reports/2013-14-legislative-session.html'
```

Set a request to grab content from the url and set it up as a BeautifulSoup object

```python
# downloads the webpage and stores it in the result object
result= requests.get(url)

# all the content of the webpage is stores in the object c
c= result.content
```

```python
#set as Beautiful Soup object
soup = BeautifulSoup(c)
```

```python
print (soup.prettify())
```

```python
#select the element at a top level
list(soup.children) 
```

Use inspect element on the webpage to find the name of the content that you want to scrap. So BeautifulSoup can be used to search the table and grab the content

```python
summary = soup.find('div', {'class':'list-land', 'id':'content'})

#Find all the tables in the webpage
tables = summary.find_all('table')

tables
```

Use BeautifulSoup to find all the entries of the table, each cell

```python
# Set rows as first indexed object in tables with a row
rows = tables[0].findAll('tr')

rows
```

```python
# Set up empty data list
data = []

# now grab every HTML cell in every row
for tr in rows:
    cols = tr.findAll('td')
    # Check to see if text is in the row
    for td in cols:
        text = td.find(text=True) 
        print (text)
        data.append(text)
```

```python
 data
```

Only get the links with pdf and date

```python
# Set up empty lists
reports = []
date = []
```

```python
index = 0

for item in data:
    if 'pdf' in item:
        date.append(data[index - 1])
        reports.append(item.replace(u'\xa0', u' '))
    index += 1
```

```python
reports
```

Changing the list to DataFrame

```python
date = Series(date)
reports = Series(reports)
```

```python
legislative_df = pd.concat([date, reports], axis = 1)
```

```python
legislative_df.columns = ('Date', 'Reports')
```
